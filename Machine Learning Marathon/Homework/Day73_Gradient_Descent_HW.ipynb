使用function y=(x+5)² 來練習學習率( leanrning rate ) 調整所造成的影響

# 作業目標:
    請嘗試使用不同的組合驗證 learning rate 對所需 iteration 的影響
    lr = [0.1, 0.0001]
    主要驗證 Lr 對於grandient 收斂的速度
    

```python
cur_x = 3 # The algorithm starts at x=3
precision = 0.000001 #This tells us when to stop the algorithm
max_iters = 10000 # maximum number of iterations
df = lambda x: 2*(x+5) #Gradient of our function 
```

### Learning Rate = 0.0001

```python
'''
需要執行初始值設定, 下列三個
# Learning rate
#iteration counter
#Step size  
'''
cur_x = 3
lr = 0.0001
iters = 1
tolerance = 1
x_history_lr1 = []
iteration_lr1 = []

# while tolerance > precision and iters < max_iters:
while iters < max_iters:
    prev_x = cur_x
    cur_x = cur_x - lr* df(prev_x)
    tolerance = abs(cur_x - prev_x)
    iters+=1
    if iters%100 == 0:
        print("the learning rate is",lr, "\nIteration",iters,"\nX value is",cur_x) #Print iterations
    x_history_lr1.append(prev_x)
    iteration_lr1.append(iters)
```

### Learning Rate = 0.001

```python
cur_x = 3
lr = 0.001
iters = 1
tolerance = 1
x_history_lr2 = []
iteration_lr2 = []

# while tolerance > precision:# and iters < max_iters:
while iters < max_iters:
    prev_x = cur_x
    cur_x = cur_x - lr * df(prev_x)
    tolerance = abs(cur_x - prev_x)
    iters+=1
    if iters%100 == 0:
        print("the learning rate is",lr, "\nIteration",iters,"\nX value is",cur_x) #Print iterations
    x_history_lr2.append(prev_x)
    iteration_lr2.append(iters)### Learning Rate = 0.005
```

### Learning Rate = 0.1

```python
cur_x = 3
lr = 0.1
iters = 1
tolerance = 1
x_history_lr3 = []
iteration_lr3 = []

# while tolerance > precision:# and iters < max_iters:
while iters < max_iters:
    prev_x = cur_x
    cur_x = cur_x - lr * df(prev_x)
    tolerance = abs(cur_x - prev_x)
    iters+=1
    if iters%50 == 0:
        print("the learning rate is",lr, "\nIteration",iters,"\nX value is",cur_x) #Print iterations
    x_history_lr3.append(prev_x)
    iteration_lr3.append(iters)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize = (15,8))
plt.plot(iteration_lr1, x_history_lr1, 'r', label='lr = 0.0001')

plt.plot(iteration_lr2, x_history_lr2, 'b', label='lr = 0.001')

plt.plot(iteration_lr3, x_history_lr3, 'g', label='lr = 0.1')
plt.legend(loc=1,fontsize=16)
plt.xlabel('iteration times',fontsize=16)
plt.ylabel('x',fontsize=16)
```
